```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df,train_size = 0.8,test_size = 0.2, random_state = 293)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_train.head()
```

```python
df_train = pd.DataFrame(scaler.fit_transform(df_train), columns=df_train.columns)
df_test = pd.DataFrame(scaler.transform(df_test),columns=df_train.columns)
```

```python
y_train = df_train.pop('Research')
x_train = df_train
y_test = df_test.pop('Research')
x_test = df_test
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
#lets fit all those features into the Logistic regression model 
lr = LogisticRegression()
lr.fit(x_train,y_train)

#doing the prediction on test data
y_pred = lr.predict(x_test)

from sklearn.metrics  import classification_report
print(classification_report(y_test, y_pred,digits=4))
```

```python
from sklearn.decomposition import PCA
```

```python
#PCA 
pca_low = PCA(0.90)
df_train_pca_low = pca_low.fit_transform(x_train)
df_train_pca_low.shape
```

```python
# Training the Regression Model
learner_pca_low = LogisticRegression()
model_pca_lr_low = learner_pca_low.fit(df_train_pca_low,y_train)
df_test_pca_low = pca_low.transform(x_test)
df_test_pca_low.shape
```

```python
#Making prediction on the test data
pred_proba_test_low = model_pca_lr_low.predict_proba(df_test_pca_low)[:,1]
"{:2.2f}".format(metrics.roc_auc_score(y_test, pred_proba_test_low))
```

```python
def calculate_Pca(percentage = .75):
    #pca on 90 percent 
    pca_low = PCA(percentage)
    df_train_pca_low = pca_low.fit_transform(x_train)
    # Training the Regression Model
    learner_pca_low = LogisticRegression()
    model_pca_lr_low = learner_pca_low.fit(df_train_pca_low,y_train)
    df_test_pca_low = pca_low.transform(x_test)
    #Making prediction on the test data
    pred_proba_test_low = model_pca_lr_low.predict_proba(df_test_pca_low)[:,1]

    return df_train_pca_low.shape,df_test_pca_low.shape,"{:2.4f}".format(metrics.roc_auc_score(y_test, pred_proba_test_low))
```

```python
for x in np.arange(0.1,1,.1):    
    print(f"percentage value = {x}")
    train_pca_shape,test_pca_shape,score  = calculate_Pca(x)
    print(train_pca_shape,test_pca_shape,score)
    print()
```
